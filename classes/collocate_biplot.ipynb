# Biplots for collocates

```python
import pandas as pd
```

Here's some made-up collocaton data:

```python
x = {'love':0.1,
    'hate':0.2,
    'serenity':0.8,
    'distraction':0.3,
    'boredom':0.5,
    'disgust':0.9,
    'surprise':0.5}
```

```python
y = {'boredom':0.9,
    'disgust':0.7,
    'surprise':0.5,
    'terror':0.1,
    'love':0.6,
    'hate':1.0,
    'distraction':0.8}
```

Two things to notice about this data: `x` and `y` are only partially overlapping. `boredom` only appears in x, and `terror` only appears in y. 

This is how most collocation data will be: some appear in both, some appear in only one. We can find words in both list using a Python `set`.

A pandas `series` object is the same as a data frame column:

```python
x = pd.Series(x)
```

```python
y = pd.Series(y)
```

```python
def collocate_biplot(x, y, x_name, y_name, figsize = (12,12)):
    '''Assumes x and y are pandas series objects with labeled indexes.'''
    import matplotlib.pyplot as plt # import our plotting library using the plt convention
    
    # we're going to use a set to find out which words are in both:
    intersection = sorted(list(set(x.index) & set(y.index)))
    
    # then filter both values for those points
    x = x[intersection]
    y = y[intersection]
    
    # make figure
    fig, ax = plt.subplots()
    ax.scatter(x, y, c = 'b')
    ax.grid(True, linestyle = ':', linewidth=0.5)

    # label points
    for i, txt in enumerate(intersection):
        ax.annotate(txt, (x[i], y[i]))
        
    # label axes
    ax.set_xlabel(x_name)
    ax.set_ylabel(y_name)
    
    # write figure to disk
    fig = plt.gcf()
    fig.set_size_inches(figsize)
    fn = x_name + '_' + y_name + '.png'
    fig.savefig(fn, dpi = 300)
```

```python
collocate_biplot(x, y, 'tories', 'whigs')
```

Note that by default the function saves a copy of the figure to your current working directory based on the names that you pass for the variables `x_name` and `y_name`.
